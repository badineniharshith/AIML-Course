# Understanding Tensors from 0D to 5D

Tensors are generalizations of scalars, vectors, and matrices to higher dimensions. They're crucial in areas like deep learning and physics, allowing data to be structured with varying complexity and dimensionality. The "D" in $$n$$D tensor refers to the number of dimensions (or axes) the data has.

Below, you'll find concise definitions, two examples for each type of tensor, and explanations for the relevant terminology.

## 0D Tensor (Scalar)

**Definition:**  
A 0D tensor represents a single numerical value. It has no dimensions—it is just a scalar.

**Examples:**
- The number $$7$$
- $$\pi \approx 3.1416$$

**Explanation of terms:**
- **Scalar:** A single number with no direction or axes, representing a magnitude only.

## 1D Tensor (Vector)

**Definition:**  
A 1D tensor is an array of numbers arranged in a single line. It has one axis—length.

**Examples:**
- Daily temperatures in a week: $$[30, 31, 29, 28, 32, color vector: $$[255, 128,anation of terms:**
- **Vector:** An ordered collection of numbers in a single line (like a list).

## 2D Tensor (Matrix)

**Definition:**  
A 2D tensor is a rectangular grid of numbers (rows and columns). It has two axes—height and width.

**Examples:**
- Grayscale pixel intensities in a 3x3 image:
  $$
  \begin{bmatrix}
  52 & 80 & 91 \\
  67 & 24 & 55 \\
  89 & 60 & 73
  \end{bmatrix}
  $$
- Table of sales data (products vs. months):  
  $$\begin{bmatrix} 100 & 120 & 95 \\ 80 & 85 & 90 \end{bmatrix}$$

**Explanation of terms:**
- **Matrix:** A 2D grid (rows and columns) of numbers.

## 3D Tensor

**Definition:**  
A 3D tensor extends matrices into a third axis, often visualized as a "cube" of numbers. It has depth, height, and width.

**Examples:**
- RGB image (height × width × color channels), e.g., a 32×32×3 image.
- Batch of sentences, each represented as a matrix of word embeddings: (number of sentences × sentence length × embedding size).

**Explanation of terms:**
- **Axis:** A dimension along which values can vary. A 3D tensor has three such axes.
- **Depth:** The third axis, like the number of color channels (red, green, blue).

## 4D Tensor

**Definition:**  
A 4D tensor is a collection of 3D tensors, adding another axis (often batch size or time).

**Examples:**
- A batch of RGB images: (batch size × height × width × color channels).  
  For 128 images, each 32×32 pixels with 3 channels: shape = (128, 32, 32, 3)
- Video as a sequence of images: (frames × height × width × channels).

**Explanation of terms:**
- **Batch:** Grouping multiple data samples into one tensor for efficient processing.
- **Channels:** For images, refers to color components (R, G, B).

## 5D Tensor

**Definition:**  
A 5D tensor has five axes, commonly used to represent batches of videos or sequences of 4D tensors.

**Examples:**
- A batch of videos: (batch size × frames × height × width × channels).  
  E.g., 10 videos, each 16 frames, 64x64 pixels, 3 channels: shape = (10, 16, 64, 64, 3)
- Volumetric medical images over time (patients × scans × depth × height × width).

**Explanation of terms:**
- **Frames:** Individual images in a sequence (e.g., in a video).
- **Volumetric data:** Data with three spatial dimensions (like MRI scans), possibly tracked over time or across batches.

### Table: Tensor Types Overview

| Dimension | Common Name | Example Shape    | Real-World Example                                   |
|-----------|-------------|------------------|------------------------------------------------------|
| 0D        | Scalar      | -                | Temperature: 23°C                                    |
| 1D        | Vector      | (3,)             |  (vector of 3 scores)                   |
| 2D        | Matrix      | (3, 2)           | 3x2 grayscale image or table                         |
| 3D        | Cube/Array  | (32, 32, 3)      | RGB image (height, width, channels)                  |
| 4D        | Array       | (128, 32, 32, 3) | Batch of RGB images                                  |
| 5D        | Array       | (10, 16, 64, 64, 3) | Batch of videos (batch, frames, height, width, channels) |

### Additional Terms

- **Shape:** Tuple showing size along each axis (e.g., (128, 32, 32, 3)).
- **Axis/Axes:** Direction(s) along which data is organized—the number equals the tensor’s dimensionality.
- **Batching:** Technique for efficient processing by grouping multiple data samples.

# Tensors from 0D to 5D — Tutorial for Google Colab Notebook

This guide explains tensors with dimensionality from 0D (scalar) up to 5D, showcasing how each type can be represented and used in Python with TensorFlow or PyTorch in a Google Colab notebook.

## Setup: Import Required Libraries

```python
import tensorflow as tf
import torch
```

## 0D Tensor (Scalar)

- **Definition:** A single value with no dimensions.
- **Example:**

TensorFlow:
```python
scalar_tf = tf.constant(7)
print(scalar_tf)
```

PyTorch:
```python
scalar_torch = torch.tensor(7)
print(scalar_torch)
```

## 1D Tensor (Vector)

- **Definition:** 1-dimensional array of numbers.
- **Example:**

TensorFlow:
```python
vector_tf = tf.constant([1.0, 2.0, 3.0])
print(vector_tf)
```

PyTorch:
```python
vector_torch = torch.tensor([1.0, 2.0, 3.0])
print(vector_torch)
```

## 2D Tensor (Matrix)

- **Definition:** 2D grid (rows × columns).
- **Example:**

TensorFlow:
```python
matrix_tf = tf.constant([[1, 2], [3, 4]])
print(matrix_tf)
```

PyTorch:
```python
matrix_torch = torch.tensor([[1, 2], [3, 4]])
print(matrix_torch)
```

## 3D Tensor

- **Definition:** Extends 2D matrix to an additional axis (e.g., depth).
- **Example:**

TensorFlow (RGB image shape example):
```python
tensor3d_tf = tf.constant([[[255, 0, 0], [0, 255, 0]],
                          [[0, 0, 255], [255, 255, 0]]])
print(tensor3d_tf)
print(tensor3d_tf.shape)
```

PyTorch:
```python
tensor3d_torch = torch.tensor([[[255, 0, 0], [0, 255, 0]],
                              [[0, 0, 255], [255, 255, 0]]])
print(tensor3d_torch)
print(tensor3d_torch.shape)
```

## 4D Tensor

- **Definition:** Collection of 3D tensors, common for batches of images.
- **Example:**

TensorFlow (batch of two RGB images, each 2x2 pixels):
```python
tensor4d_tf = tf.constant([
    [[[255, 0, 0], [0, 255, 0]],
     [[0, 0, 255], [255, 255, 0]]],
    [[[10, 10, 10], [20, 20, 20]],
     [[30, 30, 30], [40, 40, 40]]]])
print(tensor4d_tf)
print(tensor4d_tf.shape)
```

PyTorch:
```python
tensor4d_torch = torch.tensor([
    [[[255, 0, 0], [0, 255, 0]],
     [[0, 0, 255], [255, 255, 0]]],
    [[[10, 10, 10], [20, 20, 20]],
     [[30, 30, 30], [40, 40, 40]]]])
print(tensor4d_torch)
print(tensor4d_torch.shape)
```

## 5D Tensor

- **Definition:** Used for batches of videos or volumetric data over batches.
- **Example:**

TensorFlow (batch of 2 videos, each with 2 frames, 2x2 pixels, 3 color channels):
```python
tensor5d_tf = tf.constant([
    [
        [[[255, 0, 0], [0, 255, 0]],
         [[0, 0, 255], [255, 255, 0]]],
        [[[10, 10, 10], [20, 20, 20]],
         [[30, 30, 30], [40, 40, 40]]]
    ],
    [
        [[[5, 5, 5], [15, 15, 15]],
         [[25, 25, 25], [35, 35, 35]]],
        [[[45, 45, 45], [55, 55, 55]],
         [[65, 65, 65], [75, 75, 75]]]
    ]
])
print(tensor5d_tf)
print(tensor5d_tf.shape)
```

PyTorch:
```python
tensor5d_torch = torch.tensor([
    [
        [[[255, 0, 0], [0, 255, 0]],
         [[0, 0, 255], [255, 255, 0]]],
        [[[10, 10, 10], [20, 20, 20]],
         [[30, 30, 30], [40, 40, 40]]]
    ],
    [
        [[[5, 5, 5], [15, 15, 15]],
         [[25, 25, 25], [35, 35, 35]]],
        [[[45, 45, 45], [55, 55, 55]],
         [[65, 65, 65], [75, 75, 75]]]
    ]
])
print(tensor5d_torch)
print(tensor5d_torch.shape)
```
